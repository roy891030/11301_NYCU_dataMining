```python
import pandas as pd

data_csv = pd.read_csv('/content/drive/MyDrive/碩一上課堂/1131_dataMining/dataMining/HW3/_2021.csv')
data_csv.columns = data_csv.columns.str.strip()
data_csv.head()
```

## 1.1 取出 10.11.12 月資料

```python
data_csv['日期'] = pd.to_datetime(data_csv['日期'], errors='coerce')
data_oct_nov_dec = data_csv[data_csv['日期'].dt.month.isin([10, 11, 12])].copy()
data_oct_nov_dec
```

## 1.2 缺失值以及無效值以前後 1 小時平均值取代 (如果前 1 小時仍有空值，再取更前 1 小時)

```python
import numpy as np

# 將每小時的數據轉換為數值類型，將無效值轉換為NaN
for hour in range(24):
    data_oct_nov_dec[str(hour)] = pd.to_numeric(data_oct_nov_dec[str(hour)], errors='coerce')

# 針對每個小時欄位，逐行處理缺失值或無效值
for index, row in data_oct_nov_dec.iterrows():
    for hour in range(24):
        if pd.isnull(row[str(hour)]):
            # 定義前後一小時的初始值
            prev_hour = hour - 1
            next_hour = hour + 1
            prev_value = None
            next_value = None

            # 找到前一個有效值（如果有）
            while prev_hour >= 0:
                prev_value = row[str(prev_hour)]
                if not pd.isnull(prev_value):
                    break
                prev_hour -= 1

            # 找到後一個有效值（如果有）
            while next_hour <= 23:
                next_value = row[str(next_hour)]
                if not pd.isnull(next_value):
                    break
                next_hour += 1

            # 計算平均值以替換
            valid_values = []
            if prev_value is not None and not pd.isnull(prev_value):
                valid_values.append(prev_value)
            if next_value is not None and not pd.isnull(next_value):
                valid_values.append(next_value)
            if valid_values:
                data_oct_nov_dec.at[index, str(hour)] = np.mean(valid_values)
```

```python
# 1. 檢查是否還有缺失值
missing_values_count = data_oct_nov_dec.isnull().sum().sum()
print(f"資料集中缺失值的數量：{missing_values_count}")
```

## 1.3 NR 表示無降雨，以 0 取代

```python
rainfall_rows = data_oct_nov_dec['測項'] == 'RAINFALL'
data_oct_nov_dec.loc[rainfall_rows, [str(hour) for hour in range(24)]] = data_oct_nov_dec.loc[rainfall_rows, [str(hour) for hour in range(24)]].replace('NR', 0)
```

## 1.4 將資料切割成訓練集 (10.11月) 以及測試集 (12月)

```python
# 將數據切割成訓練集和測試集
train_data = data_oct_nov_dec[data_oct_nov_dec['日期'].dt.month.isin([10, 11])]
test_data = data_oct_nov_dec[data_oct_nov_dec['日期'].dt.month == 12]
```

## 1.5製作時序資料: 將資料形式轉換為行 (row) 代表18種屬性，欄 (column) 代表逐時數據資料

```python
# 選取訓練集的數據
train_data_filtered = train_data.copy()

# 取得屬性列表
attributes = train_data_filtered['測項'].unique()

# 初始化一個空的 DataFrame 以儲存轉換後的資料
time_series_data = pd.DataFrame()

# 將每種屬性分別提取，並將其逐小時數據合併成單一列
for attribute in attributes:
    # 選取該屬性的資料，並將其逐小時數據展開成單一列
    attribute_data = train_data_filtered[train_data_filtered['測項'] == attribute]
    hourly_data = attribute_data.iloc[:, 3:].values.flatten()  # 提取逐小時數據並展開
    time_series_data[attribute] = hourly_data  # 添加為新的 DataFrame 列

# 轉置 DataFrame 使每行代表一個屬性
time_series_data = time_series_data.T
time_series_data.columns = range(61 * 24)  # 將欄名設置為逐時數據的索引
```

```python
# 對 test_data 進行時序資料轉換，將每個屬性作為行，逐小時數據作為列
test_data_filtered = test_data.copy()
attributes = test_data_filtered['測項'].unique()

time_series_test_data = pd.DataFrame()

# 將每種屬性逐小時數據合併成單一列
for attribute in attributes:
    attribute_data = test_data_filtered[test_data_filtered['測項'] == attribute]
    hourly_data = attribute_data.iloc[:, 3:].values.flatten()
    time_series_test_data[attribute] = hourly_data

# 轉置使每行代表一個屬性，並設置為31天的逐時索引
time_series_test_data = time_series_test_data.T
time_series_test_data.columns = range(31 * 24)
```

## 模型訓練

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import xgboost as xgb

# 構建 PM2.5 和所有屬性數據的訓練與測試特徵和目標
pm25_series = time_series_data.loc['PM2.5'].values
X_pm25_1hr = np.array([pm25_series[i:i+6] for i in range(len(pm25_series) - 6)])
Y_pm25_1hr = np.array([pm25_series[i+6] for i in range(len(pm25_series) - 6)])
X_pm25_6hr = np.array([pm25_series[i:i+6] for i in range(len(pm25_series) - 11)])
Y_pm25_6hr = np.array([pm25_series[i+11] for i in range(len(pm25_series) - 11)])

# 提取所有屬性的數據
X_all_features_1hr = np.array([time_series_data.iloc[:, i:i+6].values.flatten() for i in range(len(pm25_series) - 6)])
X_all_features_6hr = np.array([time_series_data.iloc[:, i:i+6].values.flatten() for i in range(len(pm25_series) - 11)])

# 測試集構建
pm25_series_test = time_series_test_data.loc['PM2.5'].values
X_test_pm25_1hr = np.array([pm25_series_test[i:i+6] for i in range(len(pm25_series_test) - 6)])
Y_test_pm25_1hr = np.array([pm25_series_test[i+6] for i in range(len(pm25_series_test) - 6)])
X_test_pm25_6hr = np.array([pm25_series_test[i:i+6] for i in range(len(pm25_series_test) - 11)])
Y_test_pm25_6hr = np.array([pm25_series_test[i+11] for i in range(len(pm25_series_test) - 11)])

X_test_all_features_1hr = np.array([time_series_test_data.iloc[:, i:i+6].values.flatten() for i in range(len(pm25_series_test) - 6)])
X_test_all_features_6hr = np.array([time_series_test_data.iloc[:, i:i+6].values.flatten() for i in range(len(pm25_series_test) - 11)])

# 初始化模型
results = {}

# 1. Linear Regression - 單一 PM2.5 特徵
lr_pm25_1hr = LinearRegression().fit(X_pm25_1hr, Y_pm25_1hr)
lr_pm25_6hr = LinearRegression().fit(X_pm25_6hr, Y_pm25_6hr)

results['LR_PM25_1hr_MAE'] = mean_absolute_error(Y_test_pm25_1hr, lr_pm25_1hr.predict(X_test_pm25_1hr))
results['LR_PM25_6hr_MAE'] = mean_absolute_error(Y_test_pm25_6hr, lr_pm25_6hr.predict(X_test_pm25_6hr))

# 2. Linear Regression - 全部屬性特徵
lr_all_1hr = LinearRegression().fit(X_all_features_1hr, Y_pm25_1hr)
lr_all_6hr = LinearRegression().fit(X_all_features_6hr, Y_pm25_6hr)

results['LR_All_1hr_MAE'] = mean_absolute_error(Y_test_pm25_1hr, lr_all_1hr.predict(X_test_all_features_1hr))
results['LR_All_6hr_MAE'] = mean_absolute_error(Y_test_pm25_6hr, lr_all_6hr.predict(X_test_all_features_6hr))

# 3. XGBoost - 單一 PM2.5 特徵
xgb_pm25_1hr = xgb.XGBRegressor(n_estimators=50, max_depth=2).fit(X_pm25_1hr, Y_pm25_1hr)
xgb_pm25_6hr = xgb.XGBRegressor(n_estimators=50, max_depth=2).fit(X_pm25_6hr, Y_pm25_6hr)

results['XGB_PM25_1hr_MAE'] = mean_absolute_error(Y_test_pm25_1hr, xgb_pm25_1hr.predict(X_test_pm25_1hr))
results['XGB_PM25_6hr_MAE'] = mean_absolute_error(Y_test_pm25_6hr, xgb_pm25_6hr.predict(X_test_pm25_6hr))

# 4. XGBoost - 全部屬性特徵
xgb_all_1hr = xgb.XGBRegressor(n_estimators=50, max_depth=2).fit(X_all_features_1hr, Y_pm25_1hr)
xgb_all_6hr = xgb.XGBRegressor(n_estimators=50, max_depth=2).fit(X_all_features_6hr, Y_pm25_6hr)

results['XGB_All_1hr_MAE'] = mean_absolute_error(Y_test_pm25_1hr, xgb_all_1hr.predict(X_test_all_features_1hr))
results['XGB_All_6hr_MAE'] = mean_absolute_error(Y_test_pm25_6hr, xgb_all_6hr.predict(X_test_all_features_6hr))

# 格式化輸出模型結果
print("模型結果:")
for key, value in results.items():
    print(f"{key}: {value}")
```
